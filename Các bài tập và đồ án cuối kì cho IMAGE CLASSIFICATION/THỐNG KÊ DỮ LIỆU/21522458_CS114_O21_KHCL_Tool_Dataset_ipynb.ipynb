```python
import os
import pandas as pd
from collections import defaultdict
import re
```

```python

def get_stats_dict(base_path = './',
                   folders = []):
  # Dùng defaultdict lỡ folder có empty thì sẽ là 0.00
  image_count = defaultdict(lambda: defaultdict(float))

  # Tóm tắt Regex: số(?có thể connect bởi '-Số').(Chữ).số.chữ
  # Dĩ nhiên nếu muốn có thể check đoạn chữ sau nếu có phải là file ảnh không
  filename_regex = re.compile(r'(\d+(?:-\d+)*)\.(\w+)\.\d+\.\w+$')

  for folder in folders:
      folder_path = os.path.join(base_path, folder)
      if os.path.exists(folder_path) and os.path.isdir(folder_path):
          for filename in os.listdir(folder_path):
              match = filename_regex.match(filename)
              if match:
                  # lấy student & tên xe từ cái format trên!!!
                  student_ids, motorbike_type = match.groups()
                  # Split the student_ids by '-' và count
                  student_ids_splited = student_ids.split('-')
                  student_group_len = len(student_ids_splited)
                  for student_id in student_ids_splited:
                      image_count[student_id][motorbike_type] += 1 / student_group_len

  return image_count


def write_csv(base_dir='drive/MyDrive/my_shared_shortcut/Public',
              output_dir='./', folders=['Honda','Suzuki', 'Yamaha', 'VinFast', 'Others'],
              file_name_All='MotocycleDataset-1.csv',
              file_name_types='MotocycleDataset-2.csv'):

  image_count= get_stats_dict(base_path = base_dir, folders = folders  )

  data = []
  all_data = []
  for student_id, motorbike_data in image_count.items():
      all_total_images = 0
      for motorbike_type, total_images in motorbike_data.items():
          all_total_images += total_images
          data.append({'MSSV': student_id, 'Hiệu xe': motorbike_type, 'Số lượng': total_images})
      all_data.append({'MSSV': student_id, 'All': "All", 'Số lượng': all_total_images})

  df = pd.DataFrame(data)

  df_all = pd.DataFrame(all_data)

  # Export the DataFrame to a CSV file
  output_csv_path = os.path.join(output_dir, 'MotocycleDataset-2.csv')
  df.to_csv(output_csv_path, index=False)

  all_output_csv_path = os.path.join(output_dir, 'MotocycleDataset-1.csv')
  df_all.to_csv(all_output_csv_path,index=False)

  print(f'CSV file saved at: {output_csv_path}')
  print(f'ALL CSV file saved at: {all_output_csv_path}')
  return df , df_all

```

```python
base_dir='drive/MyDrive/my_shared_shortcut/Public'
output_dir='./'
folders=['Honda','Suzuki', 'Yamaha', 'VinFast', 'Others']
file_name_All='MotocycleDataset-1.csv'
file_name_types='MotocycleDataset-2.csv'
# Có thể set các value trên và pass vào nhé!!!!!!
write_csv()
```
